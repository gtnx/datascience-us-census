```python
# -*- coding: utf-8 -*-

import os
import pandas
import math
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
```

# 0. Chargement des données

On charge les données en considérant que les "?" sont des valeurs manquants.

```python
fn_train = "data/us_census_full/census_income_learn.csv"
headers = map(lambda l: l.strip("\n"), open("data/us_census_full/headers.txt").readlines())
continuous_headers = ["age", "wage per hour", "capital gains", "capital losses", "dividends from stocks", "num persons worked for employer", "weeks worked in year"]
dtype = dict([(h, np.int64 if h in continuous_headers else np.object) for h in headers])

df_train = pandas.read_csv(
    fn_train,
    names=headers,
    skipinitialspace=True,
    na_values=["?"],
    nrows=10000,
    dtype=dtype
)
df_train = pandas.DataFrame(df_train, columns=[c for c in df_train.columns if c != "instance weight"])
print(df_train["class"].value_counts() / len(df_train["class"]))
print(pandas.Series([df_train[name].dtype for name in df_train.columns[:-1]]).value_counts())
```

# 2. Audit des données

Le but de cette section est d'analyser les colonnes du dataset une à une. Le but de cet audit est de **comprendre au mieux les données**. Cela permet de se donner des idées sur d'éventuelles variables composites, anticiper des résultats aberrants etc

Quelques statistiques globales:
* 9 colonnes ont des données manquantes

```python
(pandas.core.common.isnull(df_train).sum() > 0).sum()
```

Ensuite, on va faire un audit de chaque variable indépendamment les unes des autres. L'audit consiste selon le type de variable en:
* Pour les variables réelles:
 * Un histogramme
 * Les statistiques de base: moyenne, écart-type, min, les trois quartiles, max
 * Le nombre de valeurs manquantes
* Pour les variables catégorielles:
 * Un pie chart des 7 valeurs les plus représentées
 * Le nombre de valeurs manquantes
 * Le nombre de valeurs distinctes

```python
columns = df_train.columns[:12]
ncols = 3
nrows = int(math.ceil(len(columns) / float(ncols)))
fig, axes = plt.subplots(nrows=nrows, ncols=ncols)
plt.subplots_adjust(wspace=1, hspace=.4)
def get_pos(k, ncols):
    return (k / ncols, k % ncols)

def audit_column(k, name, series):
    stats = series.describe()
    ratio = 100. * (len(series) - stats["count"]) / len(series)
       
    if series.dtype == np.int64 or series.dtype == np.float64:
        keys = [u'mean', u'std', u'min', u'25%', u'50%', u'75%', u'max']
        desc = ", ".join(["%s=%.2f" % (key, stats[key]) for key in keys])
        title = "%(name)s\n%(ratio).0f%% missing\n%(desc)s" % locals()
        series.hist(ax=axes[get_pos(k, ncols)])
    else:
        values = series.value_counts()
        value_nb = len(values)
        values[:7].plot(kind="pie", ax=axes[get_pos(k, ncols)], figsize=(21, 2*len(columns)))
        title = "%(name)s\n%(ratio).0f%% missing, %(value_nb)s distinct values" % locals()
    axes[get_pos(k, ncols)].set_title(title)

for k, name in enumerate(columns):
    audit_column(k, name, df_train[name])
```

# 4. Apprentissage

Le but de cet étape est d'apprendre un modèle de classification à partir des données disponibles. Il s'agit de trouver le modèle le plus **performant** selon une **métrique** définie en évitant l'écueil du **sur-apprentissage**. Ainsi, nous allons diviser le problème en plusieurs sous-problèmes:
* Choix d'une métrique adaptée au but final de la classification
* Optimisation des algorithmes de classification au moyen de la validation croisée
* Comparaison des algorithmes de classification

## 4.1 Choix de la métrique

Choisir la bonne métrique en adéquation au problème initial est indispensable car le choix de l'algorithme optimal peut complètement changer en fonction de la métrique. La littérature utilise régulièrement des mesures **moyennes**:
* La précision sur une des classes
* Le rappel sur une des classes
* Le F1 Score sur une des classes qui combine précision et rappel
* [AUC](http://en.wikipedia.org/wiki/Receiver_operating_characteristic) qui estime la qualité du modèle pour les deux classes

Dans un cas industriel, le modèle moyen n'est pas forcément le meilleur pour l'application. Par exemple, si le but de cette classification est de trouver des personnes fraudant sur la déclaration d'impôts (ie déclarant moins de 50k$ par an mais gagnant plus), de fournir les meilleurs candidats à des controleurs et que ces derniers peuvent traiter 10% des cas, il faut alors trouver l'algorithme ayant **la meilleure précision** sur les **10 pourcents qui appartiennent le plus probablement à la classe 1**. Cette métrique n'est pas **moyenne** et privilégiera peut-être d'autres algorithmes que le AUC.

**Néanmoins, sans plus d'informations sur l'utilisation de cette modélisation, nous utiliserons la métrique AUC.**

## 4.2 Optimisation des algorithmes de classification

La plupart des algorithmes classiques de classification dépendent de **paramètres** dont certains contrôlent la **complexité** de l'algorithme. Ces algorithmes sont prévus pour être d'autant plus précis qu'ils sont complexes mais s'exposent au risque du **sur-apprentissage**. Afin de trouver les bonnes valeurs des paramètres, il convient d'utiliser la technique de la **validation croisée**. Cette technique consiste à séparer plusieurs fois de suite les données d'entrée en données d'apprentissage et de test, construire un modèle sur les premières puis l'évaluer sur les secondes. La performance du classifieur est la moyenne des performances observées sur ces échantillonages.

L'algorithme ci-dessous effectue une validation croisée de type **Stratified K Fold**.

Nous allons tester deux algorithmes:
* AdaBoost
* Regression Logistique

### 4.2.1 AdaBoost

L'exemple ci-dessous montre l'utilisation de la validation croisée avec un algorithme de boosting: **AdaBoost**. Cette algorithme peut se voir comme la composition de **classifieur faibles**. On cherche à trouver le paramètre optimal pour une métrique de type AUC.

```python
from sklearn.ensemble import AdaBoostClassifier

features = df_train_bin.copy()
classes = features.pop("class")

plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
clf = AdaBoostClassifier(n_estimators=1)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 1")
clf = AdaBoostClassifier(n_estimators=4)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 4")
clf = AdaBoostClassifier(n_estimators=10)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 10")
clf = AdaBoostClassifier(n_estimators=40)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 40")
clf = AdaBoostClassifier(n_estimators=100)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 100")
clf = AdaBoostClassifier(n_estimators=400)
run_cv(features.values, classes.values, clf, 6, "AdaBoost 400")
```

Cet exemple montre que l'AUC maximale est atteint avec **40 classifieurs faibles** pour une **AUC de 0.94**. On voit qu'au-dessus, on s'expose à du sur-apprentissage.

Cette algorithme permet de voir également l'importance de chaque variable. Nous affichons les 10 premières variables en terme d'importance.

```python
features = df_train_bin.copy()
classes = features.pop("class")
clf1 = AdaBoostClassifier(n_estimators=40)
clf1.fit(features.values, classes.values)
df_features = pandas.DataFrame(
    sorted([(i, features.columns[f]) for f, i in enumerate(clf1.feature_importances_) if i > 0], reverse=True)[:10], 
    columns=["importance", "feature"]
)
df_features.index = df_features.pop("feature")
df_features.plot(kind="barh", y="importance", legend=False)
```

Le graphique ci-dessus montre:
* l'âge est la variable la plus explicative
* Nos deux variables calculées font partie des plus importantes
